```python
"""
crp_cls: 법인구분 : Y( 유가 ), K( 코스닥 ), N(코넥스 ), E(기타 )
crp_nm: 종목명/법인명
crp_cd: 종목코드(6자리)/고유코드(8자리)
rpt_nm: 공시구분 +보고서명 보고서명 +기타정보
rcp_no: 접수번호
flr_nm: 공시 제출인명
rcp_dt: 공시접수일자(YYYYMMDD)
rmk: 
"""

def dart_list(auth_key, company_code, start_date):
    page_no = 1
    url = "http://dart.fss.or.kr/api/search.xml?auth="+auth_key+"&crp_cd="+company_code+"&start_dt="+start_date+"&bsn_tp=A001&bsn_tp=A002&bsn_tp=A003"+"&page_no="+str(page_no)
    resultXML=urlopen(url)  #this is for response of XML
    result=resultXML.read() #Using read method
    xmlsoup=BeautifulSoup(result,'html.parser')
    total_page = int(xmlsoup.total_page.string)
    
    data = pd.DataFrame()
    for page_no in range(1, total_page+1):
        url = "http://dart.fss.or.kr/api/search.xml?auth="+auth_key+"&crp_cd="+company_code+"&start_dt="+start_date+"&bsn_tp=A001&bsn_tp=A002&bsn_tp=A003"+"&page_no="+str(page_no)
        resultXML=urlopen(url)  #this is for response of XML
        result=resultXML.read() #Using read method
        xmlsoup=BeautifulSoup(result,'html.parser')
        total_page = int(xmlsoup.total_page.string)
    
        te=xmlsoup.findAll("list")
        for t in te:
            temp=pd.DataFrame(([[t.crp_cls.string,t.crp_nm.string,t.crp_cd.string,t.rpt_nm.string,
                t.rcp_no.string,t.flr_nm.string,t.rcp_dt.string, t.rmk.string]]),
                columns=["crp_cls","crp_nm","crp_cd","rpt_nm","rcp_no","flr_nm","rcp_dt","rmk"])
            data=pd.concat([data,temp])
    data=data.reset_index(drop=True)
    return data
```
